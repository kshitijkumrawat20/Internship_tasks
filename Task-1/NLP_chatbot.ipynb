```python
import random
import json
import pickle
import numpy as np
import tensorflow as tf
import nltk
from  nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

```

```python
words = []
classes = []
documents = []
ignore_letters = ['?','!','.',',']
```

```python
for intent in intents["intents"]:
  for pattern in intent["patterns"]:
    wordlist = nltk.word_tokenize(pattern)
    words.extend(wordlist)
    documents.append((wordlist, intent["tag"]))
    if intent["tag"] not in classes:
      classes.append(intent["tag"])
```

```python
training = []
output_empty = [0] * len(classes)

for document in documents:
  bag = []
  word_patterns = document[0]
  word_patterns = [lemmatizer.lemmatize(word.lower()) for word in word_patterns]
  for word in words:
    bag.append(1) if word in word_patterns else bag.append(0) # creates bag of words

  output_row = list(output_empty)  # Create a new output_row for each document
  output_row[classes.index(document[1])] = 1

  training.append([bag, output_row]) # Append bag and output_row as separate elements

# Convert training to NumPy array
#training = np.array(training)  # No need to shuffle here

# Separate features (train_x) and labels (train_y)
train_x = [item[0] for item in training]
train_y = [item[1] for item in training]

train_x = np.array(train_x)
train_y = np.array(train_y)

# Now you can shuffle train_x and train_y together if needed
#from sklearn.utils import shuffle
#train_x, train_y = shuffle(train_x, train_y, random_state=0)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(len(train_y[0]), activation='softmax'))

sgd = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
hist = model.fit(np.array(train_x), np.array(train_y), epochs=500, batch_size=5, verbose=1)
model.save('chatbotmodel.h5', hist)
print("Done")
```
